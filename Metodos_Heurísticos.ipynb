
## Introducción a las Heurísticas

La **heurística** es una técnica de búsqueda diseñada para encontrar **buenas soluciones aproximadas** a problemas complejos, especialmente aquellos de tipo **combinatorio** o **no lineal** que no pueden resolverse eficientemente mediante métodos exactos de optimización.

Una heurística emplea **reglas prácticas** o de sentido común para **explorar el espacio de soluciones**, con el objetivo de encontrar una solución "suficientemente buena" en un tiempo razonable, aunque no garantice alcanzar el óptimo global.

### Ventajas y desventajas de las heurísticas

- **Ventajas**:
  - Encuentran soluciones **rápidamente**.
  - Requieren **menos recursos computacionales** que los métodos exactos.
  - Pueden aplicarse a problemas **mal definidos** o con información **incompleta**.

-  **Desventajas**:
  - No garantizan una solución óptima.
  - La **calidad de la solución** puede ser difícil de evaluar.
  - El rendimiento depende fuertemente del **problema específico** y de la heurística utilizada.

### Evolución de las heurísticas

Las primeras heurísticas se basaban en estrategias **codiciosas (greedy)**, que mejoraban la solución paso a paso al seleccionar la opción localmente más prometedora. Esta estrategia suele detenerse en un **óptimo local**, donde ya no es posible mejorar la solución con pequeños cambios.

En la década de 1980 surgieron las **metaheurísticas**, una nueva generación de algoritmos que incorporan mecanismos para **escapar de óptimos locales**, como la aceptación de soluciones peores en ciertas condiciones o el uso de poblaciones de soluciones. Esto permite una **exploración más amplia** del espacio de búsqueda, aunque con un mayor costo computacional.

## Heurísticas de Variable Discreta

En problemas de variable discreta, el espacio de búsqueda está formado por **valores enteros** o combinaciones **finitas** de opciones. La heurística codiciosa puede aplicarse mediante la exploración de vecinos cercanos para intentar mejorar la solución actual.

### Ejemplos comunes:
- Asignación de tareas.
- Problema del viajante (*TSP*).
- Programación de horarios.
- Selección de rutas o caminos óptimos.

Se pueden utilizar distintos tipos de vecindarios:

- **Vecindad inmediata**: se evalúan solo soluciones vecinas que difieren en una unidad o movimiento simple.
- **Vecindad extendida**: se consideran múltiples soluciones vecinas en un entorno más amplio.

Este tipo de exploración puede representarse como una búsqueda en una red o grafo, donde los nodos son soluciones y los arcos representan movimientos entre ellas.


## Heurísticas de Variable Continua

En problemas de optimización continua, las variables pueden tomar **valores reales**, y la función objetivo suele ser **suave** (diferenciable). Las heurísticas permiten aproximar el óptimo sin calcular gradientes exactos, lo cual es útil cuando la función es **costosa de evaluar** o **no tiene derivadas disponibles**.

### Técnicas típicas:
- **Búsqueda local**: explora en torno a una solución actual mediante pequeñas perturbaciones.
- **Recocido simulado** (*Simulated Annealing*): acepta soluciones peores con cierta probabilidad para escapar de óptimos locales.
- ** Búsqueda tabu**
-  **Algoritmos evolutivos**:
Los **algoritmos evolutivos** son una familia de metaheurísticas inspiradas en los principios de la **evolución natural** y la **selección darwiniana**. Trabajan con una **población de soluciones** que evoluciona a lo largo del tiempo mediante operadores como:

- **Selección**: se eligen las soluciones más aptas según la función objetivo.
- **Cruzamiento (crossover)**: se combinan dos soluciones para generar nuevas.
- **Mutación**: se introducen pequeñas modificaciones aleatorias para mantener la diversidad.

Estos algoritmos son especialmente útiles en espacios de búsqueda complejos, no lineales o con múltiples óptimos locales.

#### Algoritmos evolutivos comunes:
  - **Algoritmos genéticos** (*Genetic Algorithms*): utilizan operadores de selección, cruce y mutación sobre una población de soluciones.
  - **Estrategias evolutivas**.Las **estrategias evolutivas** se inspiran en los procesos de evolución biológica, pero se enfocan más en la **adaptación de parámetros** que en el uso de operadores como el cruce genético.

En lugar de trabajar con cadenas binarias como en los algoritmos genéticos, se manipulan directamente **vectores de números reales**
  -  **Enjambre de partículas** (*Particle Swarm Optimization*): los agentes se mueven en el espacio de soluciones imitando el comportamiento de grupos en la naturaleza.

### Aplicaciones:
- Ajuste de parámetros en modelos complejos.
- Optimización de funciones no convexas.
- Diseño óptimo en ingeniería y sistemas.




### Ejemplo de búsqueda con heurística codiciosa (greedy search): función discreta
#### Heurística codiciosa para optimización discreta

La heurística codiciosa busca una solución óptima local en un espacio discreto $S$ minimizando la función objetivo $F(x)$:

$$
\min_{x \in S} F(x)
$$

### Definiciones

- **Espacio de soluciones:** $S = \{1, 2, \dots, n\}$
- **Función objetivo:** $F : S \to \mathbb{R}$
- **Vecindad inmediata:** Para un punto $x_k \in S$, la vecindad se define como

$$
N(x_k) = \{x_k - 1, \, x_k + 1\} \cap S
$$

(Consideramos sólo puntos factibles dentro de $S$.)

### Algoritmo paso a paso

1. **Inicialización:**  
   Seleccionar un punto inicial $x_0 \in S$, por ejemplo, usando un número aleatorio.

2. **Iteración $k$:**
   - Calcular la vecindad $N(x_k)$.
   - Evaluar $F(x)$ para todos $x \in N(x_k)$.
   - Elegir el siguiente punto $x_{k+1}$ como

   $$
   x_{k+1} = \arg\min_{x \in N(x_k)} F(x)
   $$

3. **Condición de parada:**  
   Si

   $$
   F(x_k) \leq F(x), \quad \forall x \in N(x_k)
   $$

   entonces $x_k$ es un mínimo local y se detiene el algoritmo.

4. **Repetir** los pasos 2 y 3 hasta alcanzar la condición de parada.

### Limitación principal

Este método puede quedar atrapado en mínimos locales porque sólo se mueve hacia vecinos inmediatos con mejor valor, sin considerar movimientos más amplios o saltos en el espacio de soluciones.

### Ejemplo
Considere la función $F(x)$ dada en la figura y defina el problema de optimización como:

$$
\text{Minimizar } F(x), \quad x \in S = \{1, 2, 3, 4, 5, 6, 7, 8\}
$$

La función tiene un **mínimo local** en $x = 3$ (etiquetado como $B$), y un **mínimo global** en $x = 7$ (etiquetado como $D$).

La heurística usa una **vecindad inmediata**, definida como:

$$
N(x_k) = \{x_k - 1, \, x_k + 1\}
$$
El número aleatorio $R = 0.1002$ selecciona el punto de inicio $x = 1$ dentro del conjunto de puntos factibles $S = \{1, 2, \dots, 8\}$.

- En la iteración 1, la vecindad es $N(1) = \{0, 2\}$. Como $x = 0$ no es factible, se considera solo $x = 2$ con $F(2) = 60$.
- En la iteración 2, $N(2) = \{1, 3\}$. Se comparan $F(1) = 90$ y $F(3) = 50$.
- En la iteración 3, $N(3) = \{2, 4\}$. Como $F(3) = 50$ es menor que $F(2) = 60$ y $F(4) = 80$, se cumple que $F(3) < F(x)$ para todo $x \in N(3)$.

Por lo tanto, la búsqueda termina en la iteración 3, en el **mínimo local** $x^* = 3$ con $F(x^*) = 50$.

| 1             | 1                  | $\{0, 2\}$         | $\{2\}$                      | $F(2) = 60$                       | 2                                  | $x=0$ no es factible                         |

Esto ilustra cómo la búsqueda codiciosa puede quedar atrapada en un mínimo local.

2. Expandiendo el tamaño de la vecindad para alcanzar más puntos factibles de solución.

## Metaheurística

En la heurística, en cada iteración $k$, la búsqueda se desplaza hacia un nuevo punto $\mathbf{X}_{k+1} \in N(\mathbf{X}_k)$ sólo si este punto mejora el valor de la función objetivo $\mathbf{F(X)}$.  

Si no es posible encontrar un punto mejor en la vecindad $N(\mathbf{X}_k)$ o si se alcanza un límite máximo de iteraciones establecido por el usuario, la búsqueda termina, quedando atrapada en un óptimo local.

Por el contrario, las **metaheurísticas** están diseñadas para evitar este atrapamiento permitiendo, bajo ciertas condiciones, movimientos hacia soluciones con peor calidad temporalmente. Esta flexibilidad adicional busca ampliar la exploración del espacio de soluciones y, por lo tanto, mejorar la calidad de la solución final.

A diferencia de la heurística codiciosa, que siempre termina al alcanzar un óptimo local, la terminación de una metaheurística puede basarse en varios criterios, tales como:

1. Superar un número máximo de iteraciones de búsqueda.
2. Exceder un número máximo de iteraciones sin encontrar una mejora en la mejor solución conocida.
3. Que la vecindad del punto actual esté vacía o no permita movimientos factibles.
4. Alcanzar una solución cuya calidad sea satisfactoria según un umbral definido.

Existen cuatro metaheurísticas comunes: la búsqueda tabú, el recocido simulado y los algoritmos genéticos y evolutivos. Estas técnicas se diferencian principalmente en la forma en que gestionan la exploración para escapar de óptimos locales.


## Algoritmo de búsqueda tabú

La búsqueda tabú (BT) es una metaheurística que, al quedar atrapada en un óptimo local, permite realizar movimientos hacia soluciones peores con el fin de escapar de dicho óptimo.

Para evitar ciclos y la repetición constante de soluciones previamente visitadas, la búsqueda tabú mantiene una **lista tabú**: una memoria temporal de movimientos recientes que están prohibidos durante un cierto número de iteraciones (periodo de tenencia).

Cuando un movimiento se añade a la lista tabú, queda deshabilitado para futuras exploraciones durante este periodo. Al finalizar, dicho movimiento se elimina de la lista, quedando disponible nuevamente.

Este mecanismo facilita una exploración más amplia del espacio de soluciones y ayuda a evitar que la búsqueda se estanque en mínimos locales, favoreciendo la convergencia a soluciones de mejor calidad.



## Algoritmo de búsqueda tabú (BT)

La búsqueda tabú es una metaheurística diseñada para escapar de óptimos locales permitiendo movimientos hacia soluciones no necesariamente mejores, pero evitando ciclos mediante una memoria llamada **lista tabú**.

### Paso a paso del algoritmo

1. **Inicialización:**

   - Seleccionar un punto inicial $x_0 \in S$.
   - Inicializar la lista tabú $T = \emptyset$.
   - Establecer el mejor punto encontrado $x^* = x_0$.
   - Definir el tamaño máximo de la lista tabú, llamado $L$.

2. **En cada iteración $k$:**

   - Calcular la vecindad $N(x_k)$ del punto actual.
   
   - De $N(x_k)$, excluir los movimientos o soluciones prohibidas por la lista tabú $T$, salvo que cumplan un criterio de aspiración (por ejemplo, si mejoran la mejor solución conocida).
   
   - Seleccionar el siguiente punto $x_{k+1} \in N(x_k)$ que minimice la función objetivo:
   
$$x_{k+1} = \arg\min_{x \in N(x_k)} F(x), \quad \text{con } x \notin T \text{ o cumple aspiración}$$

   
   - Actualizar la lista tabú $T$ agregando el movimiento o solución que llevó a $x_{k+1}$. Si la lista supera el tamaño máximo $L$, eliminar el movimiento más antiguo.
   
   - Si $F(x_{k+1}) < F(x^*)$, actualizar la mejor solución:

     $$x^* \leftarrow x_{k+1}$$

3. **Condiciones de parada:**

   - Se alcanza un número máximo de iteraciones.
   - No se encuentra mejora en un número máximo de iteraciones consecutivas.
   - Otros criterios de convergencia o tiempo computacional.

### Resumen matemático del algoritmo

- **Lista tabú:**

  $$
  T = \{m_1, m_2, \dots, m_L\}
  $$
  
  donde cada $m_i$ representa un movimiento prohibido temporalmente.

- **Criterio de aspiración:**

  Un movimiento tabú puede ser aceptado si produce una solución mejor que la mejor solución actual, es decir:

  $$
  F(x_{k+1}) < F(x^*)
  $$



Esta estrategia permite explorar regiones del espacio de búsqueda que una heurística codiciosa descartaría, reduciendo la probabilidad de quedar atrapado en mínimos locales y mejorando la calidad de la solución final.

Este ejemplo ilustra el uso del algoritmo de Búsqueda Tabú (BT) para minimizar la función $F(x)$ con las siguientes condiciones:

- Punto inicial: $x_0 = 1$
- Periodo de permanencia tabú: $\tau = 3$
- Vecindad: $N(x_k) = \{x_k - 4, \ldots, x_k - 1, x_k + 1, \ldots, x_k + 4\} - L_k$
- Selección aleatoria desde $\{1, 2, \ldots, 8\}$ según un valor $R_k$ dado
- Los elementos tabú no se consideran en la vecindad

$$
\text{Parámetros:} \quad q = 4, \quad \tau = 3
$$

### Parámetros

- Punto inicial: $x_0 = 1$
- Período de permanencia tabú: $\tau = 3$
- Vecindad: $N(x_k) = \{x_k - 4, \ldots, x_k - 1, x_k + 1, \ldots, x_k + 4\} \cap \{1, \ldots, 8\}$
- Se excluyen los elementos de la lista tabú $L_k$
- Se usa un valor aleatorio $R_k \in [0, 1)$ para seleccionar un vecino:  
  Si $N(x_k)$ tiene $n$ elementos, se selecciona el vecino en el índice $i = \lfloor R_k \cdot n \rfloor$

$$
\text{Parámetros:} \quad q = 4, \quad \tau = 3
$$

### Valores aleatorios predefinidos:

$$
R_0 = 0.0935,\quad R_1 = 0.4128,\quad R_2 = 0.2039,\quad R_3 = 0.0861,\quad R_4 = 0.5839,\quad R_5 = 0.5712,\quad R_6 = 0.9489
$$


### Iteración por Iteración

#### Iteración 0 (Inicio)

- $x_0 = 1$, $F(x_0) = 90$, $R_0 = 0.0935$
- Vecindad: $N(x_0) = \{2, 3, 4, 5\}$
- Lista tabú: $L_0 = \emptyset$
- Selección aleatoria: 4 elementos → índice $\lfloor 0.0935 \times 4 \rfloor = 0$ → se elige $x_1 = 2$
- Lista tabú actualizada: $L_1 = \{1\}$

#### Iteración 1

- $x_1 = 2$, $F(x_1) = 60$, $R_1 = 0.4128$
- $N(x_1) = \{1, 3, 4, 5, 6\} \setminus \{1\} = \{3, 4, 5, 6\}$
- Índice $\lfloor 0.4128 \times 4 \rfloor = 1$ → se elige $x_2 = 4$
- $L_2 = \{1, 2\}$

#### Iteración 2

- $x_2 = 4$, $F(x_2) = 80$, $R_2 = 0.2039$
- $N(x_2) = \{1, 2, 3, 5, 6, 7, 8\} \setminus \{1, 2\} = \{3, 5, 6, 7, 8\}$
- Índice $\lfloor 0.2039 \times 5 \rfloor = 1$ → se elige $x_3 = 5$
- $L_3 = \{1, 2, 4\}$

#### Iteración 3

- $x_3 = 5$, $F(x_3) = 100$, $R_3 = 0.0861$
- $N(x_3) = \{1, 2, 3, 4, 6, 7, 8\} \setminus \{1, 2, 4\} = \{3, 6, 7, 8\}$
- Índice $\lfloor 0.0861 \times 4 \rfloor = 0$ → se elige $x_4 = 3$
- $L_4 = \{2, 4, 5\}$

#### Iteración 4

- $x_4 = 3$, $F(x_4) = 50$, $R_4 = 0.5839$
- $N(x_4) = \{1, 2, 4, 5, 6, 7\} \setminus \{2, 4, 5\} = \{1, 6, 7\}$
- Índice $\lfloor 0.5839 \times 3 \rfloor = 1$ → se elige $x_5 = 6$
- $L_5 = \{4, 5, 3\}$

#### Iteración 5

- $x_5 = 6$, $F(x_5) = 40$, $R_5 = 0.5712$
- $N(x_5) = \{2, 3, 4, 5, 7, 8\} \setminus \{4, 5, 3\} = \{2, 7, 8\}$
- Índice $\lfloor 0.5712 \times 3 \rfloor = 1$ → se elige $x_6 = 7$
- $F(x_6) = 20$ → **mejor solución encontrada**
