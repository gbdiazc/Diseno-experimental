## Vectores

El conjunto ${\mathbb{R}}^n$ es el conjunto de todas las *n*-uplas de números reales.

En notación de conjuntos, esto se expresa como:  
${\mathbb{R}}^n = \{(x_1, x_2, x_3, \cdots, x_n) : x_1, x_2, x_3, \cdots, x_n \in {\mathbb{R}}\}$.

Por ejemplo, el conjunto ${\mathbb{R}}^3$ representa el conjunto de ternas reales, es decir, las coordenadas $(x, y, z)$ en el espacio tridimensional.

Un **vector** en ${\mathbb{R}}^n$ es una *n*-upla, o punto, en ${\mathbb{R}}^n$. Los vectores pueden escribirse de forma horizontal (es decir, con los elementos del vector uno al lado del otro) como un **vector fila**, o de forma vertical (es decir, con los elementos del vector uno debajo del otro) como un **vector columna**. Si el contexto de un vector no es claro, por lo general se asume que es un vector columna.

El $i$-ésimo elemento de un vector $v$ se denota como $v_i$. La transpuesta de un vector columna es un vector fila de la misma longitud, y viceversa. En matemáticas, la transpuesta se denota con un superíndice $T$, es decir, $v^T$.

El **vector cero** es el vector en ${\mathbb{R}}^n$ que contiene únicamente ceros.

La **norma** de un vector es una medida de su longitud. Existen varias formas de definir la longitud de un vector dependiendo de la métrica utilizada (es decir, la fórmula de distancia que se elige). La más común se denomina **norma $L_2$**, y se calcula de acuerdo con la fórmula de distancia que probablemente conoces desde la escuela.

- La **norma $L_2$** de un vector $v$ se denota como $\Vert v \Vert_{2}$ y se define por:  
  $$\Vert v \Vert_{2} = \sqrt{\sum_i v_i^2}.$$  
  Esta norma también se conoce como *longitud euclidiana* y representa la longitud "física" del vector en uno, dos o tres dimensiones.

- La **norma $L_1$**, o *Distancia Manhattan*, se calcula como:  
  $$\Vert v \Vert_{1} = \sum_i |v_i|.$$  
  Recibe este nombre por la estructura en forma de cuadrícula de las calles de la ciudad de Nueva York.

- La **p-norma** o **norma $L_p$** de un vector se define como:  
  $$\Vert v \Vert_{p} = \sqrt[p]{\sum_i v_i^p}.$$

- La **norma $L_\infty$** es la $p$-norma en el caso en que $p = \infty$. Se escribe como $\Vert v \Vert_{\infty}$ y corresponde al valor absoluto máximo entre los elementos del vector $v$.



```python
import numpy as np
vector_row = np.array([[1, -5, 3, 2, 4]])
vector_column = np.array([[1], 
                          [2], 
                          [3], 
                          [4]])
print(vector_row.shape)
print(vector_column.shape)
```

```python
# Definimos un vector como arreglo de NumPy
v = np.array([3, -4, 1])
```

```python
# Norma L2 (euclidiana)
norma_L2 = np.linalg.norm(v, ord=2)
print(f"Norma L2: {norma_L2}")
```

```python

# Norma L1 (distancia Manhattan)
norma_L1 = np.linalg.norm(v, ord=1)
print(f"Norma L1: {norma_L1}")
```

```python
# Norma L∞ (máximo valor absoluto)
norma_Linf = np.linalg.norm(v, ord=np.inf)
print(f"Norma L∞: {norma_Linf}")
```

```python
# Norma Lp con p = 3
p = 3
norma_Lp = np.linalg.norm(v, ord=p)
print(f"Norma L{p}: {norma_Lp}")
```

Nota: La norma $L_\infty$ de un vector ** s igual al valor máximo** de los elementos del vector (el mayor valor sin considerar el signo).


## Suma, multiplicación de vectores y espacios vectoriales

### **Suma de vectores**

La **suma de vectores** se define como la suma componente a componente de cada uno de los elementos de los vectores que se suman. Por ejemplo, si $v$ y $w$ son vectores en ${\mathbb{R}}^n$, entonces $u = v + w$ está definida como:

$$
u_i = v_i + w_i
$$

### **Multiplicación de vectores**

La **multiplicación de vectores** puede definirse de distintas maneras dependiendo del contexto.

#### **Multiplicación escalar**

La **multiplicación escalar** de un vector es el producto de un vector por un **escalar** (es decir, un número real en ${\mathbb{R}}$). Esta se define como el producto de cada elemento del vector por el escalar. Más específicamente, si $\alpha$ es un escalar y $v$ es un vector, entonces:

$$
u = \alpha v \quad \text{está definida como} \quad u_i = \alpha v_i
$$

Nota: esta es exactamente la forma en que Python implementa la multiplicación escalar de un vector.

Por la suma de vectores, si $u = v + w$ entonces $x = \alpha u$ da:

$$
x_i = \alpha(v_i + w_i)
$$

Como $\alpha$, $v_i$ y $w_i$ son escalares, la multiplicación distribuye:

$$
x_i = \alpha v_i + \alpha w_i \quad \Rightarrow \quad \alpha(v + w) = \alpha v + \alpha w
$$


### **Producto punto (dot product)**

El **producto punto** de dos vectores es la suma del producto de los elementos correspondientes de cada vector. Se denota con un punto central: $v \cdot w$ (se lee “v punto w”). Si $v$ y $w$ son vectores en ${\mathbb{R}}^n$, el producto punto se define como:

$$
d = v \cdot w = \sum_{i=1}^{n} v_i w_i
$$

El **ángulo entre dos vectores**, $\theta$, está dado por la fórmula:

$$
v \cdot w = \|v\|_2 \|w\|_2 \cos{\theta}
$$

El producto punto mide cuán alineados están dos vectores. Por ejemplo, los vectores $(1,1)$ y $(2,2)$ son paralelos. Si se calcula el ángulo entre ellos usando el producto punto, se obtiene que $\theta = 0$. Si el ángulo $\theta = \pi/2$, los vectores son **ortogonales** (perpendiculares) y el producto punto es cero.


### **Producto cruz (cross product)**

El **producto cruz** entre dos vectores $v$ y $w$ se denota como $v \times w$ y está definido como:

$$
v \times w = \|v\|_2 \|w\|_2 \sin{(\theta)} \, \mathbf{n}
$$

donde:
- $\theta$ es el ángulo entre $v$ y $w$ (que se puede obtener del producto punto),
- $\mathbf{n}$ es un vector unitario (de longitud 1) perpendicular tanto a $v$ como a $w$.

**Interpretación geométrica:** El producto cruz genera un nuevo vector perpendicular a los dos vectores originales, cuya magnitud es igual al área del paralelogramo definido por $v$ y $w$.


## **Espacios vectoriales**

Los **espacios vectoriales** son un concepto abstracto en matemáticas. Un espacio vectorial es un conjunto $V$ de elementos llamados **vectores**, con operaciones de **suma** y **multiplicación escalar** definidas que cumplen las siguientes propiedades (donde $u$, $v$ y $w$ son vectores arbitrarios de $V$, y $c$ y $d$ son escalares):

### **Axiomas de cerradura**

- La suma $u + v$ existe y pertenece a $V$ (cerradura bajo la suma).
- El producto escalar $cu$ pertenece a $V$ (cerradura bajo la multiplicación escalar).

### **Axiomas de la suma**

- $u + v = v + u$ (propiedad conmutativa).
- $u + (v + w) = (u + v) + w$ (propiedad asociativa).
- Existe un elemento $0 \in V$ (vector cero) tal que $u + 0 = u$.
- Para cada $u \in V$, existe un $-u$ (opuesto aditivo) tal que $u + (-u) = 0$.

### **Axiomas de la multiplicación escalar**

- $c(u + v) = cu + cv$
- $(c + d)u = cu + du$
- $c(du) = (cd)u$
- $1u = u$


### Cálculo del ángulo entre dos vectores

Dados dos vectores en $\mathbb{R}^3$:

$$
v = [10, 9, 3] \quad \text{y} \quad w = [2, 5, 12]
$$

#### Producto punto (dot product)

El producto punto entre $v$ y $w$ está definido como:

$$
v \cdot w = \sum_{i=1}^3 v_i w_i = v_1 w_1 + v_2 w_2 + v_3 w_3
$$

Para nuestros vectores:

$$
v \cdot w = 10 \times 2 + 9 \times 5 + 3 \times 12 = 20 + 45 + 36 = 101
$$


#### Norma (longitud) de un vector

La norma $L_2$ de un vector $v$ es:

$$
\|v\| = \sqrt{\sum_{i=1}^3 v_i^2} = \sqrt{v_1^2 + v_2^2 + v_3^2}
$$

Calculando para $v$ y $w$:

$$
\|v\| = \sqrt{10^2 + 9^2 + 3^2} = \sqrt{100 + 81 + 9} = \sqrt{190} \approx 13.784
$$

$$
\|w\| = \sqrt{2^2 + 5^2 + 12^2} = \sqrt{4 + 25 + 144} = \sqrt{173} \approx 13.153
$$


#### Definición del ángulo entre vectores

El ángulo $\theta$ entre dos vectores se define usando la fórmula:

$$
v \cdot w = \|v\| \|w\| \cos \theta
$$

Despejando $\theta:$

$$
\theta = \arccos\left(\frac{v \cdot w}{\|v\| \|w\|}\right)
$$


#### Cálculo numérico del ángulo

Sustituyendo los valores:

$$
\theta = \arccos\left(\frac{101}{13.784 \times 13.153}\right) = \arccos\left(\frac{101}{181.3}\right) = \arccos(0.557)
$$

Calculando el arcocoseno (en radianes):

$$
\theta \approx 0.9799 \text{ radianes}
$$

Convertido a grados:

$$
\theta \approx 0.9799 \times \frac{180}{\pi} \approx 56.15^\circ
$$




```python
import numpy as np
from numpy.linalg import norm
from numpy import arccos, dot, degrees
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definición de vectores
v = np.array([10, 9, 3])
w = np.array([2, 5, 12])

# Cálculo del ángulo
theta = arccos(dot(v, w) / (norm(v) * norm(w)))

# Configurar figura 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Origen
origin = np.zeros(3)

# Graficar vectores
ax.quiver(*origin, *v, color='blue', label='v = [10, 9, 3]', arrow_length_ratio=0.1)
ax.quiver(*origin, *w, color='red', label='w = [2, 5, 12]', arrow_length_ratio=0.1)

# Para mostrar el ángulo, graficamos una pequeña curva en el plano formado por v y w
# Primero normalizamos los vectores
v_norm = v / norm(v)
w_norm = w / norm(w)

# Generar puntos para el arco del ángulo
angle_steps = np.linspace(0, theta, 100)
arc_points = np.array([np.cos(t) * v_norm + np.sin(t) * (w_norm - np.dot(w_norm, v_norm) * v_norm) for t in angle_steps])
arc_points /= norm(arc_points, axis=1)[:, np.newaxis]  # Normalizar cada punto
arc_points *= 3  # Escalar para que el arco sea visible

# Graficar el arco del ángulo
ax.plot(arc_points[:, 0], arc_points[:, 1], arc_points[:, 2], color='green', label=f'Ángulo θ ≈ {degrees(theta):.2f}°')

# Ajustes estéticos
ax.set_xlim([0, max(v[0], w[0]) + 2])
ax.set_ylim([0, max(v[1], w[1]) + 2])
ax.set_zlim([0, max(v[2], w[2]) + 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Visualización de vectores y el ángulo entre ellos')

plt.show()
```

Dado los vectores $$ v = [0, 2, 0] $$ y $$ w = [3, 0, 0] $$, el producto cruz $$ v \times w $$ se calcula como:

$$
v \times w =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
0 & 2 & 0 \\
3 & 0 & 0 \\
\end{vmatrix}
= \mathbf{i}(2 \cdot 0 - 0 \cdot 0) - \mathbf{j}(0 \cdot 0 - 0 \cdot 3) + \mathbf{k}(0 \cdot 0 - 2 \cdot 3)
$$

$$
v \times w = \mathbf{i}(0) - \mathbf{j}(0) + \mathbf{k}(-6) = [0, 0, -6]
$$

Por lo tanto,

$$
\boxed{v \times w = [0, 0, -6]}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir vectores
v = np.array([0, 2, 0])
w = np.array([3, 0, 0])

# Calcular producto cruz
cross = np.cross(v, w)

print("Producto cruz v x w =", cross)

# Graficar vectores
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

origin = np.zeros(3)

# Graficar vector v
ax.quiver(*origin, *v, color='blue', label='v = [0, 2, 0]', arrow_length_ratio=0.1)
# Graficar vector w
ax.quiver(*origin, *w, color='red', label='w = [3, 0, 0]', arrow_length_ratio=0.1)
# Graficar producto cruz
ax.quiver(*origin, *cross, color='green', label='v × w = [0, 0, -6]', arrow_length_ratio=0.1)

# Ajustes de la gráfica
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 3])
ax.set_zlim([-7, 1])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.legend()
ax.set_title('Vectores v, w y su producto cruz v × w')

plt.show()
```

## Combinación lineal y ejemplo con vectores

Sea un conjunto $S$ donde están definidos la suma y la multiplicación por escalares. Una **combinación lineal** de los elementos de $S$ se define como

$$
\sum \alpha_i s_i,
$$

donde cada $\alpha_i$ es un número real (escalar) y cada $s_i$ es el $i^{\text{ésimo}}$ elemento de $S$. A veces, los valores $\alpha_i$ se llaman **coeficientes** de $s_i$. Las combinaciones lineales se usan para describir muchas cosas; por ejemplo, una cuenta de supermercado se puede escribir como

$$
\sum c_i n_i,
$$

donde $c_i$ es el costo del artículo $i$ y $n_i$ es la cantidad del artículo $i$ comprado. Así, el total es una combinación lineal de los artículos.

Un conjunto se llama **linealmente independiente** si ninguno de sus elementos puede escribirse como combinación lineal de los demás. En caso contrario, el conjunto es **linealmente dependiente**.


Dados los vectores fila

$$
v = [0, 3, 2], \quad w = [4, 1, 1], \quad u = [0, -2, 0],
$$

se quiere expresar el vector

$$
x = [-8, -1, 4]
$$

como combinación lineal de $v$, $w$ y $u$, es decir,

$$
x = \alpha v + \beta w + \gamma u,
$$

donde $\alpha, \beta, \gamma \in \mathbb{R}$ son los coeficientes a encontrar.


### Expandiendo por componentes

Por componentes, la combinación lineal es

$$
x = [x_1, x_2, x_3] = \alpha [0, 3, 2] + \beta [4, 1, 1] + \gamma [0, -2, 0].
$$

Esto equivale a las ecuaciones

$$
\begin{cases}
x_1 = 0 \cdot \alpha + 4 \cdot \beta + 0 \cdot \gamma = 4\beta, \\
x_2 = 3 \cdot \alpha + 1 \cdot \beta - 2 \cdot \gamma, \\
x_3 = 2 \cdot \alpha + 1 \cdot \beta + 0 \cdot \gamma = 2\alpha + \beta.
\end{cases}
$$

Sustituyendo $x = [-8, -1, 4]$, obtenemos el sistema

$$
\begin{cases}
-8 = 4\beta, \\
-1 = 3\alpha + \beta - 2\gamma, \\
4 = 2\alpha + \beta.
\end{cases}
$$



### Resolviendo el sistema

1. De la primera ecuación:

$$
4 \beta = -8 \implies \beta = -2.
$$

2. Sustituyendo $\beta = -2$ en la tercera ecuación:

$$
4 = 2 \alpha - 2 \implies 2 \alpha = 6 \implies \alpha = 3.
$$

3. Sustituyendo $\alpha = 3$ y $\beta = -2$ en la segunda ecuación:

$$
-1 = 3(3) + (-2) - 2 \gamma \implies -1 = 9 - 2 - 2 \gamma \implies -1 = 7 - 2 \gamma.
$$

Restando 7 a ambos lados:

$$
-8 = -2 \gamma \implies \gamma = 4.
$$


### Resultado final

Los coeficientes son

$$
\boxed{
\alpha = 3, \quad \beta = -2, \quad \gamma = 4.
}
$$

Por lo tanto, el vector $x$ se puede escribir como

$$
x = 3v - 2w + 4u.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir vectores
v = np.array([0, 3, 2])
w = np.array([4, 1, 1])
u = np.array([0, -2, 0])
x = np.array([-8, -1, 4])

# Matriz de coeficientes (v, w, u como columnas)
A = np.column_stack((v, w, u))

# Resolver sistema A * [alpha, beta, gamma] = x
coef = np.linalg.solve(A, x)
alpha, beta, gamma = coef

print(f"Coeficientes:\nα = {alpha:.2f}, β = {beta:.2f}, γ = {gamma:.2f}")

# Graficar vectores
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

origin = np.zeros(3)

# Graficar vectores v, w, u
ax.quiver(*origin, *v, color='blue', label='v = [0, 3, 2]', arrow_length_ratio=0.1)
ax.quiver(*origin, *w, color='red', label='w = [4, 1, 1]', arrow_length_ratio=0.1)
ax.quiver(*origin, *u, color='orange', label='u = [0, -2, 0]', arrow_length_ratio=0.1)

# Graficar vector x
ax.quiver(*origin, *x, color='green', label='x = [-8, -1, 4]', arrow_length_ratio=0.1)

# Ajustes de los ejes para mejor visualización
ax.set_xlim([-8, 8])
ax.set_ylim([-3, 3])
ax.set_zlim([-4, 4])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Vectores y combinación lineal')

plt.show()
```

## Ejercicio: Independencia lineal de vectores

¿El siguiente conjunto de vectores es linealmente independiente?: 

$$
v = [1, 1, 0], \quad w = [1, 0, 0], \quad u = [0, 0, 1].
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir los vectores
v = np.array([1, 1, 0])
w = np.array([1, 0, 0])
u = np.array([0, 0, 1])

# Crear figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Origen para los vectores
origin = np.array([0, 0, 0])

# Graficar vectores
ax.quiver(*origin, *v, color='r', label='v = [1,1,0]')
ax.quiver(*origin, *w, color='g', label='w = [1,0,0]')
ax.quiver(*origin, *u, color='b', label='u = [0,0,1]')

# Configurar límites de los ejes
ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título y leyenda
ax.set_title('Vectores v, w y u en 3D')
ax.legend()

plt.show()
```

Claramente, $u$ es linealmente independiente de $v$ y $w$ porque solo $u$ tiene un elemento distinto de cero en la tercera posición. Los vectores $v$ y $w$ también son linealmente independientes porque solo $v$ tiene un elemento distinto de cero en la segunda posición. Por lo tanto, los vectores $v$, $w$ y $u$ son linealmente independientes.


## Matrices

Una **matriz** de dimensión $m \times n$ es una tabla rectangular de números que consta de $m$ filas y $n$ columnas. La norma de una matriz puede considerarse como un tipo particular de norma vectorial. Si tratamos los $m \times n$ elementos de una matriz $M$ como los elementos de un vector en un espacio de dimensión $mn$, entonces la norma $p$-ésima se puede escribir como:

$$
\| M \|_{p} = \sqrt[p]{\sum_{i=1}^m \sum_{j=1}^n |a_{ij}|^p}
$$

La suma de matrices y la multiplicación por escalares funcionan igual que para vectores. Sin embargo, la **multiplicación de matrices** entre dos matrices, $P$ y $Q$, está definida cuando $P$ es una matriz de dimensión $m \times p$ y $Q$ es una matriz de dimensión $p \times n$. El resultado de $M = PQ$ es una matriz $M$ de dimensión $m \times n$. La dimensión con tamaño $p$ se llama la **dimensión interna** de la multiplicación de matrices, y debe coincidir (es decir, el número de columnas de $P$ debe ser igual al número de filas de $Q$) para que la multiplicación esté definida. Las dimensiones $m$ y $n$ se llaman las **dimensiones externas**.

Formalmente, si $P$ es $m \times p$ y $Q$ es $p \times n$, entonces el producto $M = PQ$ está definido como:

$$
M_{ij} = \sum_{k=1}^p P_{ik} Q_{kj}
$$

El producto de dos matrices $P$ y $Q$ en Python se realiza con el método **dot** de Numpy. 

La **transpuesta** de una matriz es el intercambio de sus filas por sus columnas. La transpuesta se denota con un superíndice $T$, es decir, $M^T$ es la transpuesta de la matriz $M$. En Python, se usa el atributo **T** de un arreglo de Numpy para obtener la transpuesta. Por ejemplo, si $M$ es una matriz, entonces $M.T$ es su transpuesta.


### Ejemplo

Sea la matriz

$$
M = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix},
$$

entonces su transpuesta es

$$
M^T = \begin{bmatrix}
1 &4 \\
2 & 5  \\
3 & 6 
\end{bmatrix},
$$


```python
import numpy as np

M = np.array([[1, 2, 3],
              [4, 5, 6]])

print(M)
M_transpuesta = M.T
print(M_transpuesta)
```

## Matrices cuadradas, determinantes y matriz identidad

Una **matriz cuadrada** es una matriz de tamaño ${n} \times {n}$; es decir, tiene el mismo número de filas que de columnas. Una propiedad importante de las matrices cuadradas es el **determinante**. El determinante se denota por $det(M)$ o a veces por $|M|$. Más adelante se mostrarán algunos usos del determinante.

### Determinante de una matriz $2 \times 2$

Para una matriz de tamaño $2 \times 2$:

$$
|M| = \begin{bmatrix}
a & b \\
c & d
\end{bmatrix} = ad - bc
$$

### Determinante de una matriz $3 \times 3$

Para una matriz de tamaño $3 \times 3$:

$$
|M| = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}
= aei + bfg + cdh - ceg - bdi - afh
$$

Podemos usar un enfoque similar para calcular el determinante de matrices de mayor dimensión.


## Matriz identidad

La **matriz identidad** es una matriz cuadrada con unos en la diagonal principal y ceros en el resto. Se denota con la letra $I$ y es análoga al número 1 en la multiplicación. Es decir, si multiplicamos cualquier matriz $M$ por la matriz identidad $I$ (de tamaño compatible), obtenemos la misma matriz $M$.




```python
import numpy as np

M = np.array([[0, 2, 1, 3],
              [3, 2, 8, 1],
              [1, 0, 0, 3],
              [0, 3, 2, 1]])
M
```

```python
I = np.eye(4)
I
```

```python
producto = np.dot(M, I)
print(producto)
```

```python
det_M = np.linalg.det(M)
print(f"Determinante de M: {det_M:.2f}")
```

La **inversa** de una matriz cuadrada $M$ es una matriz del mismo tamaño, $N$, tal que $M \cdot N = I$. La inversa de una matriz es análoga a la inversa de los números reales. Por ejemplo, la inversa de 3 es $\frac{1}{3}$ porque $(3)(\frac{1}{3}) = 1$. Se dice que una matriz es **invertible** si tiene una inversa. La inversa de una matriz es única; es decir, para una matriz invertible, existe una sola inversa para esa matriz. Si $M$ es una matriz cuadrada, su inversa se denota como $M^{-1}$ en matemáticas, y se puede calcular en Python utilizando la función `inv` del paquete `linalg` de Numpy.

Para una matriz de $2 \times 2$, la solución analítica de la matriz inversa es:

$$
M^{-1} = \begin{bmatrix}
a & b \\
c & d\\
\end{bmatrix}^{-1} = \frac{1}{|M|}\begin{bmatrix}
d & -b \\
-c & a\\
\end{bmatrix}
$$

El cálculo de la matriz inversa mediante la solución analítica se vuelve más complicado al aumentar la dimensión de la matriz. Existen muchos otros métodos que pueden facilitar este cálculo, como la eliminación de Gauss, el método de Newton, la descomposición en valores propios (eigendecomposition), entre otros.

Recordemos que el 0 no tiene inversa multiplicativa en el conjunto de los números reales. De manera similar, existen matrices que no tienen inversa. Estas matrices se llaman **singulares**. Las matrices que sí tienen inversa se llaman **no singulares**.

Una forma de determinar si una matriz es singular es calcular su determinante. Si el determinante es 0, entonces la matriz es singular; en caso contrario, la matriz es no singular.

La matriz $M$ (del ejemplo anterior) tiene un determinante distinto de cero. Calcula la inversa de $M$. Demuestra que la matriz 

$$
P = \begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 0 \\
1 & 0 & 1 \\
\end{bmatrix}
$$

tiene un valor de determinante igual a 0, y por lo tanto, no tiene inversa.


```python
import numpy as np

# Definir la matriz
M = np.array([[4, 7],
              [2, 6]])

# Calcular el determinante
det = np.linalg.det(M)
print(f"Determinante de M: {det}")


# Calcular la inversa
M_inv = np.linalg.inv(M)
print("Matriz inversa de M:")
print(M_inv)

# Verificar que M * M_inv = Identidad
identidad = M @ M_inv
print("Producto M * M^{-1} (debe ser la identidad):")
print(np.round(identidad))  
```

```python
from numpy.linalg import inv, det
# Matriz P que es singular
P = np.array([
    [0, 1, 0],
    [0, 0, 0],
    [1, 0, 1]
])
# Calcular el determinante de P
det_P = det(P)
print("\nDeterminante de la matriz P:")
print(det_P)

# Verificar si tiene inversa
if det_P == 0:
    print("La matriz P es singular y no tiene inversa.")
else:
    P_inv = inv(P)
    print("La inversa de P es:")
    print(P_inv)
```

### Matriz simétrica

Una matriz $A$ es **simétrica** si es igual a su transpuesta:

$$
A = A^T
$$

Esto significa que $a_{ij} = a_{ji}$ para todos los elementos. Las matrices simétricas tienen autovalores reales y autovectores ortogonales. Muchas matrices de covarianza o Hessianas en Machine Learning son simétricas.


## Autovalores y autovectores

Si $A \mathbf{v} = \lambda \mathbf{v}$, entonces $\mathbf{v}$ es un **autovector** de la matriz $A$ y $\lambda$ es su **autovalor**.

Estos conceptos son fundamentales en algoritmos como PCA y en el análisis de sistemas dinámicos.

```python
A = np.array([[1, 2], [3, 4]])
eigvals, eigvecs = np.linalg.eig(A)
print("Autovalores:", eigvals)
print("Autovectores:\n", eigvecs)
```

```python
# Crear figura
plt.figure(figsize=(6, 6))
colors = ['r', 'b']

# Graficar los autovectores escalados por sus autovalores
origin = np.array([[0, 0], [0, 0]])  # origen para flechas

for i in range(len(eigvals)):
    vec = eigvecs[:, i]
    val = eigvals[i]
    plt.quiver(*origin[:, i], *vec, angles='xy', scale_units='xy', scale=1, color=colors[i], label=f"$\\lambda_{i+1} = {val:.2f}$")

# Configuración del gráfico
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Autovectores de la matriz $A$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Matriz positiva definida

Una matriz simétrica $A$ es **positiva definida** si:

$$
\mathbf{x}^T A \mathbf{x} > 0 \quad \text{para todo } \mathbf{x} \ne 0
$$

Propiedades:
- Todos sus autovalores son positivos.
- Aparece frecuentemente como matriz Hessiana en problemas de minimización convexa.
- Si $A$ es positiva definida, la función cuadrática $f(\mathbf{x}) = \mathbf{x}^T A \mathbf{x}$ es estrictamente convexa.

```python
import numpy as np

# Matriz simétrica positiva definida
A = np.array([[2, -1],
              [-1, 2]])

# Lista de vectores no nulos
vectors = [
    np.array([1, 0]),
    np.array([0, 1]),
    np.array([1, 1]),
    np.array([1, -1]),
    np.array([2, 3])
]

# Verificar que x^T A x > 0 para cada vector
for x in vectors:
    xtAx = x.T @ A @ x
    print(f"x = {x}, x^T A x = {xtAx}")
    if xtAx > 0:
        print("Positivo\n")
    else:
        print("No es positivo\n")
```

```python
eigvals, eigvecs = np.linalg.eig(A)
print("Autovalores:", eigvals)
print("Autovectores:\n", eigvecs)
```

```python
# Crear figura
plt.figure(figsize=(6, 6))
colors = ['r', 'b']

# Graficar los autovectores escalados por sus autovalores
origin = np.array([[0, 0], [0, 0]])  # origen para flechas

for i in range(len(eigvals)):
    vec = eigvecs[:, i]
    val = eigvals[i]
    plt.quiver(*origin[:, i], *vec, angles='xy', scale_units='xy', scale=1, color=colors[i], label=f"$\\lambda_{i+1} = {val:.2f}$")

# Configuración del gráfico
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Autovectores de la matriz $A$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Una matriz simétrica $A$ es:

- **Positiva definida** si: $\mathbf{x}^T A \mathbf{x} > 0$ para todo $\mathbf{x} \ne 0$.
- **Negativa definida** si: $\mathbf{x}^T A \mathbf{x} < 0$ para todo $\mathbf{x} \ne 0$.

Esto equivale a que **todos los autovalores de $A$ sean positivos o negativos**, respectivamente.
